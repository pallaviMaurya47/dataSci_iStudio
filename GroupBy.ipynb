# Merging and Joining: 

## 4types of joining-left ,right,inner and outer

```python
div_3=pd.DataFrame({"roll_no":[1,2,3,4],"first_name":["riya","siya","keshav","sangeeta"],"last_name":["x","y","z","alpha"],"English":[10,49,50,38]})
div_4=pd.DataFrame({"roll_no":[4,5,6,7],"first_name":["sangeeta","kiran","saumya","KRISHNA"],"last_name":["alpha","AY","AZ","AA"],"Hindi":[14,39,27,40]})
```

```python
div_3
```

```python
div_4
```

```python
pd.merge(div_3,div_4,on=["roll_no","first_name","last_name"],how="outer")
```

```python
pd.merge(div_3,div_4,on=["roll_no","first_name","last_name"],how="outer")
```
