#List comprehension in python is a concise way to create a new list or dict based on the existing iterable object( like on List[],dict but not on tuple( immutability), set{unordered})

```python
#list comprehension

#n=11 natural numbers inside the list
list_natural= [i for i in range(11)]
print(list_natural)
```

```python
#list_comprehension with conditional statements: 

#even or odd numbers
num_list=[ i for i in range(12) if i%2==0]
print(num_list)
```

#Dict comprehension

```python
#creation of dictionary w/o comprehension:
list_d=[2,3,4,5,6]
dict_d={}

for i in list_d:
    dict_d[i]=i**2
print(dict_d)
```

```python
#program in dict_compre
ans_dict={i:i**2 for i in list_d}
print(ans_dict)
```

```python
#sum of numbers
sum=0
for i in range(1,11):
    sum+=i
    print(sum)

#how to do sum with comprehension
```

Pickup n natural numbers and place that inside a list,after that find the numbers which are less than 5 and perform the cube of that number and place the number as the key in dict & cube as the value in dict.

```python
n=int(input("enter the value of n: "))
n_list=[i for i in range(1,n)]
n_list
```

```python
dict_cube={ i:i**3 for i in n_list if i<5}
print(dict_cube)
```
