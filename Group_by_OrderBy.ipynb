### let have find how many drink sold today?
then use Aggregate functions: 5 
#### sum(), avg(), min(),max(),

```python
#what is the total wine serving for the asia continent:
df.groupby("continent").agg( {"wine_servings": "sum"})
```

```python
df.groupby("continent").get_group("AS").agg( {"wine_servings": "sum"})
```

```python
#sum & avg:
df.groupby("continent").agg( {"wine_servings": ["sum","mean"], "spirit_servings":"min"})
```

```python
df.dtypes
```

```python
#every column:
df.groupby("continent").sum()
```
