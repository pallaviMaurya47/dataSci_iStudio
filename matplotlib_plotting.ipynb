```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#line graph
x=[5,2,3,5,8]
y=[9,7,4,5,2]
plt.Figure(figsize=(20,7))
plt.plot(x,y,marker="o")
```

```python
plt.xlabel("Profit")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.plot(y,x,"o-.k",mec="r")

#plt.plot(y,x,"o:k",mec="r") for ..
```

### stacked plots:

```python
x=["a","b","c","d"]
y_2=[20,30,25,45]
y=[20,40,60,30]

plt.bar(x,y,color="r")
```
