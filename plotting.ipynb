```python
import seaborn as sns
```

```python
import pandas as pd
df1=pd.read_csv("D:/studymaterial/DATA SCIENCE/loan_data_set.csv")
df1
```

```python
#any feature of the details should be one of numericl type of data during plotting the graph , in either univariate or bivariate data

#univariate categorical feature:
sns.catplot(data=df1,y="Dependents",kind="count",palette="Purples")
```

```python
#only two graphs can be used in seaborn as kind=count and pie
sns.catplot(data=df1,x="Gender",kind="box")
```

```python
help(sns.catplot(data=df1))
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid") #borderlines for the pie
plt.figure(figsize=(6,6)) #enhancing the size of pie
plt.pie(df1["Gender"].value_counts(),labels=["female","male"],autopct="%1.2f%%")#plotting the pie
plt.legend()
```

```python
#univariate-numerical:distributed plot,categorical:catplot
#bivariate-numerical: pair and join plot, cat:catplot
```

# univariate numerical data:

```python
plt.figure(figsize=(10,7))
sns.distplot(df1["LoanAmount"],kde=False)
```

```python
sns.distplot(df1["LoanAmount"],hist=False) #histogram to disappear
```

```python
sns.histplot(df1["LoanAmount"],palette="Purples") #not every graph cover the color palette
```

# Bivariate numerical data:

```python
#like heatmap , join plot is
plt.figure(figsize=(100,10)) #can't enhance or reduce the joinplot because of axisgrid ; means accurate point on the graph
sns.jointplot(data=df1,x="CoapplicantIncome",y="LoanAmount",kind="scatter",hue="Dependents")
```

```python
#hue= to judge the coordinate
sns.pairplot(data=df1)
#for plot to all data : pair plot
```
