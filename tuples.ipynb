Tuples: immutable( once defined ,can't delete ,add or edit inside it)

*another standard sequence data type like List[ mutable]
*store heterogeneous & homogeneous data
ordered
index helps to traverse a tuple()

*Faster & lesser consumed memory than list as () 'static'
No built-in functions like List.Why? ==bcz immutable

```python
x_tuple= 1,2,3,4
x_tuple[2]
print(x_tuple)
```

```python
x_tuple[0]=0   #can't change
```

```python
tuple2=(3,7,5,2,"fruits","hey",9)

#use tuple slicing in place of indexing bcz it is immutable
tuple2[0:4:1]
tuple2[:4]
```

```python
tuple2[-2: ] #backward direction
tuple2[-5:-2]

tuple2[-2:-5 :-1]   #from -2 to -5 backward = "hey","fruits",2,5
```

tuple summry

```python
tuple2.count(5)
tuple2.index(3)
```

```python
# is reversing a tuple is possible? Yes using slicing(  not by reverse()method)
tuple2[ : :-1]
 #it's not going to change in original tuple 
```
