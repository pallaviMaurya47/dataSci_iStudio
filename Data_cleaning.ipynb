## filling all object type features with "no"

```python
df["LoanAmount"].describe()
```

giving mean & median, by which have to fill features?
As Std normal distribution, mean==median.But not here so if choose mean, will get disturbed
#### after imputation, no change into entries or feature( mean): choose 50%

```python
print(df["LoanAmount"].median())
```

```python
df["LoanAmount"]=df["LoanAmount"].fillna(df["LoanAmount"].median())
```

```python
df.isnull().sum()
```

```python
df["Loan_Amount_Term"]=df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].median())
```

```python
df["Credit_History"]=df["Credit_History"].fillna(df["Credit_History"].median())
```

```python
#total income
df["total_income"]=df["ApplicantIncome"]+df["CoapplicantIncome"]
```

mean=50%: correct data ,not biased

```python
#total income <5000 : low income category
#ti>5000 and t<10000 :good ,avg
# >10000 :high

def income(var):
    if var<5000:
        return "low income group"
    elif var>5000 and var<10000:
        return "mid income group"
    else:
        return "high income group"
```

```python
df["Binning"]=df["total_income"].apply(income)
```

```python
df
```

```python
#column for stability or instability
#married,graduate and not self_employed ---> stable
#else unstable
```

```python
def stable(var):
    if (var["Married"]=="Yes") & (var["Education"]=="Graduate") & (var["Self_Employed"]=="No"):
        return "Stable"
    else:
        return "Unstable"
```

```python
df["Stability"]=df.apply(stable,axis=1)  #column axis=1 
```
